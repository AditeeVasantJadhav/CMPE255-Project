```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Read dataset obtained from previous step where subjectivity_score, flesch_index and polarity are calculated

toy_data=pd.read_csv('Toys_and_Games.csv')
toy_data.head(5)
```

```python
toy_data['PreProcessed_reviewText'] = toy_data['PreProcessed_reviewText'].astype(str)
```

```python
# Add new column 'entropy_score' to dataframe

toy_data['entropy_score'] = 0
```

```python
# Compute Entropy score for each review and store value in column 'entropy_score'

import math
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
nltk.download('punkt')
```

```python
length = len(toy_data)
for i in range(length):
    review = toy_data.iloc[i,6]
    fdist = FreqDist()
    for word in word_tokenize(review):
        fdist[word.lower()] += 1
    probs = [fdist.freq(l) for l in fdist]
    toy_data.iloc[i,10] = -sum(p * math.log(p,2) for p in probs)
```

```python
# Save dataframe into same csv file so that new column will be added to it and can be used for further references

toy_data.to_csv('Toys_and_Games.csv',index=False)
```
