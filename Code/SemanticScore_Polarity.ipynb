```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Read dataset obtained from previous step where subjectivity_score and flesch_index scores are calculated

toy_data=pd.read_csv('Toys_and_Games.csv')
toy_data.head(5)
```

```python
toy_data['PreProcessed_reviewText'] = toy_data['PreProcessed_reviewText'].astype(str)
```

```python
# Add new column 'polarity_score' to dataframe

toy_data['polarity_score'] = 0
```

```python
# Compute Polarity score for each review and store value in column 'polarity_score'

from textblob import TextBlob
```

```python
length = len(toy_data)
for i in range(length):
  review = toy_data.iloc[i,6]
  polarityScore = TextBlob(review).sentiment.polarity
  toy_data.iloc[i,9] = polarityScore
```

```python
# Save dataframe into same csv file so that new column will be added to it and can be used for further references

toy_data.to_csv('Toys_and_Games.csv',index=False)
```
