```python
import pandas as pd
import numpy as np
```

```python
toy_data=pd.read_csv('Toys_and_Games.csv')
toy_data.head(5)
```

```python
toy_data['helpful_ratio'] = 0
```

```python
toy_data.head(5)
```

```python
reviewer_map={}

def calculate_votes(row):
    helpfulstr = row['helpful']
    helpfulstr = helpfulstr.strip('[]')
    helpfullist = helpfulstr.split(',')
    helpfullist = [int(num) for num in helpfullist]
    #print(helpfullist)
    reviewerID = row['reviewerID']
    if reviewerID in reviewer_map.keys():
        reviewer_map[reviewerID][0] += helpfullist[0]
        reviewer_map[reviewerID][1] += helpfullist[1]
    else:
        reviewer_map[reviewerID] = helpfullist
        

for index,row in toy_data.iterrows():
    calculate_votes(row)
```

```python
reviewer_map
```

```python
def set_helpful_ratio(reviewer_id):
    helpfullist = reviewer_map[reviewer_id]
    helpfullist[0] += 1
    helpfullist[1] += 2
    helpfulratio = helpfullist[0]/helpfullist[1]
    return helpfulratio
```

```python
toy_data['helpful_ratio'] = toy_data['reviewerID'].apply(set_helpful_ratio)
```

```python
toy_data.to_csv('Toys_and_Games.csv',index=False)
```
