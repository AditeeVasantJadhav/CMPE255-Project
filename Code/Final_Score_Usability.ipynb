```python
########################################################################################################
# IMPORT LIBRARIES
########################################################################################################
import numpy as np
import pandas as pd
```

<font color='green'>**Read dataset obtained from preprocessing step**</font>

```python
########################################################################################################
# Read dataset obtained from normalization step
########################################################################################################
toy_data=pd.read_csv('Toys_and_Games.csv')
toy_data.head(5)
```

<font color='green'>**Add new column 'usability_score' to dataframe**</font>

```python
########################################################################################################
# Add new column 'usability_score' to dataframe
########################################################################################################
toy_data['usability_score']=0
```

<font color='green'>**Compute Usability score for each review and store value in column 'usability_score'**</font>

```python
########################################################################################################
# Compute Usability score for each review and store value in column 'usability_score'
########################################################################################################
length=len(toy_data)
for i in range(length):
    subjectivityScore=toy_data.iloc[i,7]
    polarityScore=toy_data.iloc[i,15]
    entropy=toy_data.iloc[i,16]
    fleschIndex=toy_data.iloc[i,14]
    daleChallIndex=toy_data.iloc[i,17]
    usabilityScore=(subjectivityScore+polarityScore-entropy-fleschIndex+daleChallIndex)/5
    toy_data.iloc[i,18]=usabilityScore
```

<font color='green'>**Save dataframe into same csv file so that new column will be added to it and can be used for further references**</font>

```python
########################################################################################################
# Save dataframe into same csv file so that new column will be added to it and can be used for further references
########################################################################################################
toy_data.to_csv('Toys_and_Games.csv',index=False)
```
