```python
with open('EasyWords.txt') as fp:
    easywordlist = fp.readlines()
    easywordlist = [word.strip(' \n') for word in easywordlist]
```

```python
import re

def sentence_length(text):
    sentences = re.split('[!.?]',text)
    sentences = [sentence for sentence in sentences if re.search('[a-zA-Z]',sentence)]
    #print(sentences)
    return len(sentences)

def word_length(text):
    words = re.split('[^0-9A-Za-z]',text)
    words = [word for word in words if re.search('[a-zA-Z]',word)]
    #print(words)
    return len(words)

def difficult_words(text):
        words = re.split('[^0-9A-Za-z]',text)
        words = [word.lower() for word in words if re.search('[a-zA-Z]',word)]
        diff_words_text = set()
        for word in words:
            if word not in easywordlist and len(word) > 2:
                diff_words_text.add(word)
        #print(diff_words_text)
        return(len(diff_words_text))
        
        
def dale_chall_index(text):
    average_sentence_length = word_length(text)/sentence_length(text)
    diff_words_count = difficult_words(text)
    word_count = word_length(text)
    per_diff_words = (diff_words_count / word_count) * 100
    raw_score = (0.1579 * per_diff_words) + (0.0496 * average_sentence_length)
    #print(per_diff_words,diff_words_count,word_count)
    if per_diff_words > 5:
        raw_score += 3.6365
    #print(raw_score)
    return raw_score
```

```python
#sentence_length('hi all. how are you? I think its awesome! right.')
#word_length('hi all. how are you?I think its awesome! right.')
dale_chall_index('hi all spruce people. how are you?I think its good! right.')
#difficult_words('That is Wicked.')
```

```python
toy_data['DaleChall_Index'] = toy_data['reviewText'].apply(dale_chall_index)
```
