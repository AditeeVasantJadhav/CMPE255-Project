```python
#Computing Lex Diversity Ratio for each preprocessed review test using nltk library
from nltk import FreqDist
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize
import nltk
import nltk.data
```

```python
raw_data=open('C:/Users/khurd/Documents/255/Project/Toys_and_Games.csv').read()
raw_data[0:500]
```

```python
with open('C:/Users/khurd/Documents/255/Project/Toys_and_Games.csv', 'r') as reader:
    moby_raw = reader.read()
```

```python
moby_tokens = nltk.word_tokenize(moby_raw)
text = nltk.Text(moby_tokens)
moby_series = pd.Series(moby_tokens)
```

```python
#Function to Count the total number of words
def countToken():
     
    return len(moby_tokens)

MOBY_TOKEN_COUNT = countToken()
print("Dataset has {:,} words.".format(
     MOBY_TOKEN_COUNT))
```

```python
#Function to count total number of unique words
def uniqueToken():
    
    return len(set(nltk.word_tokenize(moby_raw)))

MOBY_UNIQUE_COUNT = uniqueToken()
print("Dataset has {:,} unique words.".format(
    MOBY_UNIQUE_COUNT))
```

```python
#Function for lexical diversity
def lexical_diversity(tokens):
   
    return len(set(tokens))/float(len(tokens))
```

```python
def outputLexDiversity():
    
    return lexical_diversity(moby_tokens)

output = outputLexDiversity()
print("Lexical Diversity of Dataset: {:.2f}".format(output))
```

```python
#Calculation the lexDiversity for review text
toy_data['Lex Diversity Ratio']=toy_data['PreProcessed_reviewText'].apply(lexical_diversity)
```
