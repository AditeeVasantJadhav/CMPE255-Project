## Since the minimum and maximum value of Dale Chall Index do not lie in between 0 and 1, normalization is needed. Following are the steps for it.

```python
########################################################################################################
# Add new column 'normalized_DaleChall_Index' to dataframe
########################################################################################################
toyData['normalized_DaleChall_Index'] = 0
```

```python
########################################################################################################
# Use MinMaxScaler to scale the Dale Chall Index Value between 0 t0 1
########################################################################################################
scaler = MinMaxScaler()
toyData['normalized_DaleChall_Index'] = scaler.fit_transform(toyData[['DaleChall_Index']])
```
